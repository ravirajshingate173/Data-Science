```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic_data = pd.read_csv('tested.csv')
titanic_data.head(5)
```

```python
titanic_data.info()
```

```python
titanic_data.isna().sum()
```

```python
sns.countplot(data=titanic_data, x='Sex', hue='Survived')
```

```python
numeric_cols = titanic_data.select_dtypes(include=['float64','int64']).columns
correlation_matrix = titanic_data[numeric_cols].corr()
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=1, linecolor='blue')
plt.title('Correlation Matrix of Numeric Features')
plt.show()
```
